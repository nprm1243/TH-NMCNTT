```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("books_new_N2.csv")
data.head()
```

```python
genre = data['Genre']
genre.describe()
```

```python
your_bins=5 # unique

colors = ['green', 'blue', 'lime' ,'red','purple']
labels = ['tech','science','nonfiction','fiction','philosophy']

density,bins,arr= plt.hist(genre,bins=your_bins,edgecolor='white', linewidth=1,label=labels)
i=0
count = []
for x,y in zip(bins,density):
    count.append(arr[i].get_height())
```

```python
pair = []
dem = []
Genre = data["Genre"]
SubGenre = data["SubGenre"]
for i in range(len(Genre)):
    tmp = [Genre[i], SubGenre[i]]
    if (tmp in pair):
        dem[pair.index(tmp)] += 1
    else:
        pair.append(tmp)
        dem.append(1)
        
Genre = []
SubGenre = []
for item in pair:
    Genre.append(item[0])
    SubGenre.append(item[1])
    
df = pd.DataFrame({"Genre" : Genre, "SubGenre" : SubGenre, "Count" : dem})
```

```python
freq = df
Cout = freq['Count']
```

```python
rl_data= data[['Genre','SubGenre']]
import seaborn as sns

color_dict = dict({'tech':'brown',
                  'science':'green',
                  'nonfiction': 'orange',
                  'fiction': 'red',
                   'philosophy': 'dodgerblue'})
df = freq
plt.figure(figsize=(15,8))
sns.set_style("whitegrid")
a =sns.scatterplot(data=df,x='Genre',y='SubGenre',size=Cout,sizes=(40, 400),legend='full',palette=color_dict)
a.figure.savefig("scatterplot.png",dpi=1200)
```
